```python
import urllib.request 
import xlsxwriter
from pprint import pprint 
# for parsing all the tables present 
from html_table_parser import HTMLTableParser 
# for converting the parsed data in a pandas dataframe 
import pandas as pd
```

```python
def url_get_contents(url):
    req = urllib.request.Request(url=url)
    f = urllib.request.urlopen(req) 
    #reading contents of the website 
    return f.read() 
```

```python
list_links=['https://www.financialexpress.com/market/stock-market/split-of-face-value/',
           'https://www.financialexpress.com/market/stock-market/rights-issues/',
           'https://www.financialexpress.com/market/stock-market/bonus-issues/',
           'https://www.financialexpress.com/market/stock-market/dividend-details/']
sheet_name= ['Split','right','bonus','dividend']
split_df = pd.DataFrame()
right_df = pd.DataFrame()
bonus_df = pd.DataFrame()
div_df = pd.DataFrame()
for col,name in zip(list_links,sheet_name):
    if name=='Split':
        xhtml = url_get_contents(col).decode('utf-8')
        p = HTMLTableParser() 
        p.feed(xhtml) 
        df= pd.DataFrame(p.tables[1])
        split_df = split_df.append(df)
           
    elif name=='right':
        xhtml = url_get_contents(col).decode('utf-8')
        p = HTMLTableParser() 
        p.feed(xhtml) 
        df1= pd.DataFrame(p.tables[1])
        right_df = right_df.append(df1)
    elif name=='bonus':
        xhtml = url_get_contents(col).decode('utf-8')
        p = HTMLTableParser() 
        p.feed(xhtml) 
        df2= pd.DataFrame(p.tables[1])
        bonus_df = bonus_df.append(df2)
    else:
        xhtml = url_get_contents(col).decode('utf-8')
        p = HTMLTableParser() 
        p.feed(xhtml) 
        df3 = pd.DataFrame(p.tables[1])
        #pprint(pd.DataFrame(p.tables[1]))
        div_df = div_df.append(df3)
    
```
